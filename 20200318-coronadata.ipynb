```python
import wget
import pandas as pd
import numpy as np
```

```python
url = "https://covid.ourworldindata.org/data/full_data.csv"
dataset = wget.download(url)
dataset
```

```python
data = pd.read_csv(dataset)
```

```python
data.head()
```

```python
from datetime import datetime
```

```python
data["ts"] = data.apply(lambda row: datetime.strptime(row.date, "%Y-%m-%d"), axis = 1)
```

```python
data.head()
```

```python
data["rank"] = data.groupby("location")["ts"].rank(ascending = True)
```

```python
data.head(n = 40)
```

```python
data[data["rank"] == 1.0]
```

```python
print(data[data["rank"] == 1.0])
```
