```python
def clone_name_if_na(value1, value2):
    if pd.isna(value2):
        return value1
    else:
        return value2
```

```python
tmp_2["country_2"] = tmp_2.apply(lambda row: clone_name_if_na(row.country_1, row.country_2), axis = 1)
tmp_2["country_3"] = tmp_2.apply(lambda row: clone_name_if_na(row.country_1, row.country_3), axis = 1)
```

```python
tmp_2["same_name_1"] = tmp_2.apply(lambda row: True if row.country_1 == row.country_2 else False, axis = 1)
tmp_2["same_name_2"] = tmp_2.apply(lambda row: True if row.country_1 == row.country_3 else False, axis = 1)
tmp_2["same_name_3"] = tmp_2.apply(lambda row: True if row.country_2 == row.country_3 else False, axis = 1)
```

```python
def get_latest_pop(val1, val2, val3, val4, val5):
    if not pd.isna(val5):
        return val5
    elif pd.isna(val5) and not pd.isna(val4):
        return val4
    elif pd.isna(val5) and pd.isna(val4) and not pd.isna(val3):
        return val3
    elif pd.isna(val5) and pd.isna(val4) and pd.isna(val3) and not pd.isna(val2):
        return val2
    elif pd.isna(val5) and pd.isna(val4) and pd.isna(val3) and pd.isna(val2) and not pd.isna(val1):
        return val1
    else:
        return None
```

```python
tmp_2["pop"] = tmp_2.apply(lambda x: get_latest_pop(x.pop_2016_1, x.pop_2016_2, x.pop_2017, x.pop_2018, x.pop_2019), axis = 1)
```
